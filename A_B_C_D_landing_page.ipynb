```python
import pandas as pd
df=pd.read_csv('/content/four_sessions.csv')
print(df.head())
```

```python
df['Page'].value_counts()
```

```python
df[df['Page']=='Page 1'].describe()
df[df['Page']=='Page 2'].describe()
df[df['Page']=='Page 3'].describe()
df[df['Page']=='Page 4'].describe()
```

```python
conversion_rates = df.groupby('Page')['Time'].mean()
print(conversion_rates)
```

```python
import numpy as np

page1_time = df[df['Page'] == 'Page 1']['Time']
page2_time = df[df['Page'] == 'Page 2']['Time']
page3_time = df[df['Page'] == 'Page 3']['Time']
page4_time = df[df['Page'] == 'Page 4']['Time']

counts = np.array([len(page1_time), len(page2_time), len(page3_time), len(page4_time)])
means = np.array([page1_time.mean(), page2_time.mean(), page3_time.mean(), page4_time.mean()])
counts, means
```

```python
from scipy.stats import f_oneway

# Perform ANOVA test
f_statistic, p_value = f_oneway(page1_time, page2_time, page3_time, page4_time)

print(f"F-statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.3f}")

if p_value < 0.05:
    print("There is a statistically significant difference between the mean times on the pages.")
else:
    print("There is no statistically significant difference between the mean times on the pages.")
```
